```python
#Loading the Dataset
Dataset =  pd.read_csv('Alexa_dataset.csv')
X = Dataset.iloc[:, [2, 3]].values
y = Dataset.iloc[:, 4].values
print(X)
print(y)
```

1.3: Splitting the dataset into the Training set and Test set

```python
#Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
print(X_train)
print(y_train)
print(X_test)
print(y_test)
```

1.4: Feature Scaling

**Feature Scaling**

In this step, we perform feature scaling to standardize the input features.

Why is Feature Scaling Needed?

Our input features:

Age

Minutes of Music Consumed

are measured on very different scales. For example:

Age may range from 18 to 60.

Minutes of music consumed may range from 5,000 to 50,000.

If we do not scale the data:

Features with larger values (like Minutes of Music Consumed) will dominate the learning process.

Distance-based algorithms such as KNN and SVM will give biased results.

Gradient-based algorithms (like Logistic Regression) may converge slower.

Feature scaling ensures:

Each feature contributes equally.

The model trains more efficiently.

Numerical stability improves.

Think of two students:

* Student A scores between **20–30**
* Student B scores between **200–300**

Even if both improve by the same percentage, Student B’s numbers look much larger.

A machine learning model behaves similarly.
If one feature has much larger numbers, the model may give it more importance — even if it is not truly more important.

**What StandardScaler Does**

It:

* **Brings all features to a similar scale**
* Makes their averages equal
* Makes their variation similar

So instead of comparing:

* Age = 25
* Minutes = 40,000

The model compares:

* Age = -0.3
* Minutes = 0.1

Now both features are treated fairly.

Simple Way to Think About It

StandardScaler:

“Removes size differences so features compete fairly.”

It does **not change relationships in the data**.
It only changes the *units* so learning becomes balanced and stable.


```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
scale_X = StandardScaler()
X_train = scale_X.fit_transform(X_train)
X_test = scale_X.transform(X_test)
print(X_train)
print(X_test)
```

1.5: Fitting Decision Tree Classification to the Training set

```python
# Classifier – I (Decision Tree Classifier)
# Fitting Decision Tree Classification to the Training set
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier( criterion =  'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

Here we create the decision tree model.

**What do the parameters mean?**

criterion = 'entropy'

The tree decides how to split the data using entropy.

Entropy measures how pure or mixed a group is.

The tree tries to reduce randomness after each split.

Lower entropy = better separation.

random_state = 0

Ensures reproducible results.

Every time you run the code, you get the same tree structure.

1.6:	Predicting the Decision Tree Classifier

```python
# Predicting the Decision Tree Classifier
y_pred = classifier.predict(X_test)
print(y_pred)
print(y_test)
```

1.7: Making the confusion matrix

```python
#Making the confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

1.9: Function definition for visualization of results

```python
# Function definition for visualization of results
def Visualizer(argument1, arguement2):
  from matplotlib.colors import ListedColormap
  X_set, y_set = argument1, arguement2
  X1, X2 = np.meshgrid(np.arange(start=X_set[:,0].min()-1, stop
  =X_set[:,0].max()+1, step=0.01),
  np.arange(start=X_set[:, 1].min() - 1, stop =
  X_set[:, 1].max() + 1, step = 0.01))
  plt.contourf(X1,X2,classifier.predict(np.array([X1.ravel(),
  X2.ravel()]).T).reshape(X1.shape),
  alpha= 0.75, cmap = ListedColormap(('red', 'green')))

  plt.xlim(X1.min(), X1.max())
  plt.ylim(X2.min(), X2.max())
  for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
    c = ListedColormap(('red', 'green'))(i), label=j)
  plt.xlabel('Age')
  plt.ylabel('Minutes_of_Music_Consumed')
  plt.legend()
  plt.show()
```
